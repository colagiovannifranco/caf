```python
import pandas as pd
import numpy as np
import geopandas as gpd
import urbanpy as up
import requests
import json
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
file = 'covid-geo/CcF/Villa Hermosa (Mex)/files/mexshape/gadm36_MEX_2.shp'
mxshp = gpd.read_file(file)

shape_tabasco = mxshp[mxshp['NAME_1'] == 'Tabasco']
vh_bounds = (-93.052705,17.912656,-92.7,18.097081)
around_villahermosa = shape_tabasco.cx[vh_bounds[0]:vh_bounds[2],vh_bounds[1]:vh_bounds[3]].dissolve(by = 'NAME_1')
```

```python
dac = up.geom.gen_hexagons(resolution=8, city=around_villahermosa)
dac.shape
```

```python
dac.plot()
```

```python
pop_all = up.download.hdx_fb_population('ecuador', 'full')
pop_children = up.download.hdx_fb_population('ecuador', 'children')
pop_youth = up.download.hdx_fb_population('ecuador', 'youth')
pop_elder = up.download.hdx_fb_population('ecuador', 'elderly')
```

```python
dac_all_pop = up.geom.filter_population(pop_all, dac).rename(columns = {'population_2020':'all_pop'})
dac_children_pop = up.geom.filter_population(pop_children, dac).rename(columns = {'population':'children_pop'})
dac_youth_pop = up.geom.filter_population(pop_youth, dac).rename(columns = {'population':'youth_pop'})
dac_elderly_pop = up.geom.filter_population(pop_elder, dac).rename(columns = {'population':'elder_pop'})
```

```python
dac_all_pop.head()
```

```python
dac = up.geom.merge_shape_hex(dac, dac_all_pop, how='inner', op='intersects', agg={'all_pop': 'sum'})
dac = up.geom.merge_shape_hex(dac, dac_children_pop, how='inner', op='intersects', agg={'children_pop': 'sum'})
dac = up.geom.merge_shape_hex(dac, dac_youth_pop, how='inner', op='intersects', agg={'youth_pop': 'sum'})
dac = up.geom.merge_shape_hex(dac, dac_elderly_pop, how='inner', op='intersects', agg={'elder_pop': 'sum'})
dac['middleage_pop'] = dac['all_pop'] - dac[['children_pop','youth_pop','elder_pop']].sum(1)
```

```python
grupo = dac.groupby('hex').sum()
```

```python
grupo[grupo['all_pop']>200].index
```

```python
g = dac.plot(figsize = (20,10), column = 'all_pop', legend=True, missing_kwds={'color': 'grey'}, 
             scheme = 'percentiles', cmap = 'Spectral_r')
```

```python
for var_ in ['all_pop', 'children_pop', 'youth_pop','middleage_pop','elder_pop']:
    plt.figure()
    g = dac.plot(figsize = (20,10), column = var_, legend=True, missing_kwds={'color': 'grey'}, 
                 scheme = 'percentiles', cmap = 'Spectral_r')
    g.set_title(var_)
    figu = g.get_figure()
    figu.savefig('covid-geo/CcF/Daule (Ecuador)/accesibility/' + var_)
    plt.show()
```

```python
health_facilities = up.download.overpass_pois(bounds=around_villahermosa.total_bounds, facilities='health')
```

```python
ax = around_villahermosa.plot()
health_facilities.plot(color = 'red', ax = ax)
```

```python
health_facilities['poi_type'].value_counts()
```

```python
health_facilities.shape
```

```python
dac.shape
```

```python
# Get Hexagon centroid latitude and longitude
dac['lon'] = dac.geometry.centroid.x
dac['lat'] = dac.geometry.centroid.y
```

```python
# Get distance and indexes from nearest points 
dist, ind = up.utils.nn_search(
    tree_features = health_facilities[['lat', 'lon']].values, # Point of Interest
    query_features = dac[['lat', 'lon']].values, # Hexagon
    metric = 'manhattan' # Distance metric
)
```

```python
# Add nearest health facility index to the hexagon gdf
dac['nearest_health_facility_ix'] = ind
```

```python
def osrm_route(origin, destination, how = 'car'):  # how puede ser también foot y bicycle
    r = requests.get('http://router.project-osrm.org/route/v1/' + how + '/' + 
                     str(origin.x) + ',' + 
                     str(origin.y) +  ';' + 
                     str(destination.x) + ',' + 
                     str(destination.y) +   
                     '?overview=false')
    duration = json.loads(r.text)['routes'][0]['duration']/60  # in minutes after / 60
    distance = json.loads(r.text)['routes'][0]['distance']/1000 # in km after / 1000
    return (distance, duration)
```

```python
dist_dur = dac.apply(lambda row: osrm_route(
    origin = row.geometry.centroid,
    destination = health_facilities.iloc[row['nearest_health_facility_ix']].geometry,
    how = 'car'
    ),
    result_type='expand', axis=1
)
```

```python
dist_dur_health = dist_dur
```

```python
dac['distance_health_facility'] = dist_dur[0]
dac['duration_health_facility'] = dist_dur[1]
```

```python
g = around_villahermosa.boundary.to_crs(epsg = 3857).plot(figsize = (15,10), alpha = 0, edgecolor = 'black')
ctx.add_basemap(g, url=ctx.providers.Stamen.TonerLite)

dac.to_crs(epsg = 3857).plot(ax = g, figsize = (15,10), column='duration_health_facility', 
                             alpha = 0.6,
                             cmap='magma_r', legend=True, missing_kwds={'color': 'grey'})
g.set_title('Access to health facilities in minutes by car')
g.axis('off')
figu = g.get_figure()
figu.savefig('covid-geo/CcF/Villa Hermosa (Mex)/health_access.png')
plt.show()
```
