La siguiente funcion toma como input una base de datos como el BEAM y calcula las distancias entre las ciudades de un mismo pais 

```python
def calculate_distances_from_df(database):
    final_df = []
    for country in database['COUNTRY'].unique():
        df = database[database['COUNTRY'] == country]
        for x in df['MAIN_CITY'].unique():
            pointa = df[df['MAIN_CITY'] == x]
            country = pointa['COUNTRY'].values[0]
            citya = pointa['MAIN_CITY'].values[0]
            latlonga = pointa['point'].values[0]
            for y in df['MAIN_CITY'].unique():
                if x != y:
                    pointb = df[df['MAIN_CITY'] == y]
                    cityb = pointb['MAIN_CITY'].values[0]
                    latlongb = pointb['point'].values[0]
                    print(citya,cityb,sep = ' to ')
                    response = get_json(latlonga,latlongb,apikey)
                    data = json.loads(response)

                    distance = data['rows'][0]['elements'][0]['distance']['text']
                    duration = data['rows'][0]['elements'][0]['duration']['text']
                    duration_in_traffic = data['rows'][0]['elements'][0]['duration_in_traffic']['text']
                    datapoint = [country,citya,latlonga,cityb,latlongb,distance,duration,duration_in_traffic]
                    final_df.append(datapoint)
                else:
                    continue

    data_final = pd.DataFrame(final_df)
    data_final.columns = ['Country','City A','Point A','City B','Point B','Distance','Duration','Duration in traffic']
    return data_final
```

```python
distances = pd.read_excel('shipment/distances.xlsx', index_col = 0)
```

```python
cwpops = pd.read_excel('shipment/cwpops.xlsx', index_col = [0,1])
```

Obtengo latitudes y longitudes de las ciudades

```python
latlongs = distances[['Country','City A','Point A']].rename(
    columns = {'Point A':'latlong','City A':'City'}
).drop_duplicates('City').set_index(['Country','City']).dropna()
latlongs[['lat','long']] = latlongs['latlong'].apply(lambda x: pd.Series(x.split(','))).astype(float)
latlongs = latlongs.drop('latlong', axis = 1)
```

Joineo con latlongs y creo GeoDataFrame

```python
cwpops = cwpops.join(latlongs)

cwpops = gpd.GeoDataFrame(
    cwpops, geometry=gpd.points_from_xy(cwpops.long, cwpops.lat))
```

#### Funciones

```python
def get_wages_from_cutoff(city_index, cutoff):
    city_wage = cwpops.drop_duplicates('index1').loc[city_index]['estimated_citywage']
    distan = distances[(distances['Country'] == city_index[0])
                            &(distances['City A'] == city_index[1])
                            &(distances['parsed_duration'] <= cutoff)].drop_duplicates('City B')

    nearby_cities = [(x,y) for x,y in distan[['Country','City B']].values.tolist()]
    if len(nearby_cities) == 0:
        return city_wage
    else:
        nearby_wages = cwpops.loc[nearby_cities]['estimated_citywage'].sum()
        return nearby_wages + city_wage
```

```python
def get_countryshare_from_cutoff(city_index, cutoff, share_out = True):
    nearwages = get_wages_from_cutoff(city_index, cutoff = cutoff)
    country_wages = cwpops.groupby('Country')['estimated_citywage'].sum().loc[city_index[0]]
    share = nearwages/country_wages
    if share_out == True:
        return share
    else:
        return nearwages
```

```python
samplecities = [('Mexico','Mexicali'),
                ('Mexico','La Laguna'), 
                ('Mexico','Villahermosa'),
                ('Mexico','Hermosillo'),
               ('Promedios','Promedio de Mexico'),
               ('Promedios','Promedio de Latam')]
```

```python
shares_cutoff = pd.DataFrame(index = cwpops.index)

shares_cutoff['index1'] = shares_cutoff.index
for i in range(10):
    shares_cutoff['cutoff_' + str(i)] = shares_cutoff['index1'].apply(
                lambda x: get_countryshare_from_cutoff(x, cutoff = i))
```
